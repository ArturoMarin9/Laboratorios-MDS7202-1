1. **Diferencia entre datos de entrenamiento y validaci贸n**:

La diferencia entre los datos de entrenamiento y los datos de validaci贸n radica en su uso dentro del proceso de modelado. Los datos de entrenamiento son los que se utilizan para ajustar el modelo, es decir, para ense帽arle a identificar patrones y hacer predicciones. Por otro lado, los datos de validaci贸n no se utilizan para el ajuste directo del modelo, sino para evaluar su rendimiento durante el proceso de entrenamiento y para ajustar sus hiperpar谩metros. Una matriz de confusi贸n es una herramienta utilizada para evaluar el rendimiento de un modelo de clasificaci贸n. Se trata de una tabla que desglosa las predicciones del modelo en cuatro categor铆as: Verdaderos Positivos (TP), Falsos Positivos (FP), Verdaderos Negativos (TN) y Falsos Negativos (FN). Esta matriz permite comprender mejor c贸mo est谩 funcionando el modelo en t茅rminos de precisi贸n y recall, ayudando a identificar qu茅 tipos de errores est谩 cometiendo el modelo.

4. **M茅tricas y su interpretaci贸n**:

   - **Accuracy**:
     La precisi贸n, o accuracy, se calcula como la proporci贸n de predicciones correctas sobre el total de predicciones. La f贸rmula es $Accuracy = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{FP} + \text{TN} + \text{FN}}$. Se interpreta como la efectividad general del modelo, indicando qu茅 tan frecuentemente el modelo est谩 en lo correcto.
   
   - **Precision**:
     La precisi贸n, o precision, se calcula como la proporci贸n de verdaderos positivos sobre el total de predicciones positivas. La f贸rmula es ${Precision} = \frac{TP}{TP + FP}$. Se interpreta como la exactitud de las predicciones positivas del modelo, siendo 煤til cuando el costo de los falsos positivos es alto.
   
   - **Recall**:
     El recall se calcula como la proporci贸n de verdaderos positivos sobre el total de instancias reales positivas. La f贸rmula es $Recall = \frac{TP}{TP + FN}$. Se interpreta como la capacidad del modelo para identificar todas las instancias positivas, siendo crucial cuando es importante capturar todos los casos positivos.
   
   - **F1 Score**:
     El F1 score es la media arm贸nica entre precision y recall. La f贸rmula es $F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}$. Se interpreta como una medida equilibrada que toma en cuenta tanto la precisi贸n como el recall, siendo 煤til cuando se necesita un balance entre ambos.

   - **Aprobaci贸n de cr茅ditos de alto riesgo**:

### 1.2 Entrenamiento [3 puntos]

Ahora, vamos a entrenar los pipelines generados en los pasos anteriores. Para esto, debe realizar las siguientes tareas:

1. Separe los datos de entrenamiento en un conjunto de entrenamiento y de prueba  (la proporci贸n queda a su juicio). (vea la documentaci贸n de `train_test_split`). [1 puntos]


2. Defina un pipeline llamado `pipeline_xgboost` y otro llamado `pipeline_lightgbm`. Estos pipelines deben tener el mismo ColumnTransformer definido en la secci贸n de preprocesamiento, pero deben variar los clasificadores de acuerdo al nombre de cada pipeline. [1 puntos]

3. Entrene los pipelines. [1 punto]



**Respuesta:**

````python
#1.
# Separamos los datos en features (X) y etiquetas (y)
X = df_players.drop(columns=['label'])
y = df_players['label']

# Dividimos los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Display the shapes of the resulting datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
#2.
# Definir los clasificadores
xgboost_classifier = XGBClassifier()
lightgbm_classifier = LGBMClassifier()

# Definir los pipelines
pipeline_xgboost = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', xgboost_classifier)
])

pipeline_lightgbm = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', lightgbm_classifier)
])
````

````python
#3.
# Entrenar los pipelines
pipeline_xgboost.fit(X_train, y_train)
pipeline_lightgbm.fit(X_train, y_train)
````

### 1.3 Resultados [6 puntos]

#### Decision Tree (rbol de Decisi贸n, 'dt')

Un 谩rbol de decisi贸n es un modelo de machine learning que se utiliza tanto para clasificaci贸n como para regresi贸n. Funciona dividiendo repetidamente el conjunto de datos en subconjuntos m谩s peque帽os basados en las caracter铆sticas de los datos. Cada divisi贸n se realiza en un nodo de decisi贸n, y los resultados de las divisiones se representan en las hojas del 谩rbol.

**Funcionamiento:**
1. Se selecciona una caracter铆stica del conjunto de datos que mejor separa los datos en clases diferentes.
2. Se divide el conjunto de datos en dos subconjuntos basados en esta caracter铆stica.
3. Este proceso se repite recursivamente para cada subconjunto hasta que todos los datos en un nodo pertenezcan a la misma clase o hasta alcanzar una profundidad m谩xima del 谩rbol.

#### Random Forest (Bosque Aleatorio, 'rf')

El random forest es una t茅cnica de ensemble learning que construye m煤ltiples 谩rboles de decisi贸n y los combina para obtener una predicci贸n m谩s robusta y precisa. La idea es que m煤ltiples modelos d茅biles (los 谩rboles de decisi贸n) se combinen para formar un modelo fuerte.

**Funcionamiento:**
1. Se crea una gran cantidad de 谩rboles de decisi贸n (generalmente, varios cientos) utilizando diferentes subconjuntos aleatorios del conjunto de datos y caracter铆sticas.
2. Cada 谩rbol hace una predicci贸n individual.
3. Para una tarea de clasificaci贸n, se toma la mayor铆a de votos de las predicciones individuales de todos los 谩rboles para decidir la clase final.
4. Para una tarea de regresi贸n, se promedian las predicciones de todos los 谩rboles.

#### Extra Trees (rboles Extra, 'et')

Los Extra Trees, o Extremely Randomized Trees, son una variante de los random forests. La principal diferencia radica en c贸mo se realizan las divisiones en los nodos de decisi贸n.

**Funcionamiento:**
1. Al igual que los random forests, los Extra Trees generan varios 谩rboles de decisi贸n.
2. Sin embargo, en lugar de buscar la mejor divisi贸n en cada nodo, las divisiones se seleccionan de manera totalmente aleatoria.
3. Este proceso puede reducir la varianza del modelo y aumentar su velocidad, ya que no se requiere una b煤squeda exhaustiva de la mejor divisi贸n en cada nodo.

#### Comparaci贸n:
- **rbol de Decisi贸n (dt)**: Sencillo y f谩cil de interpretar, pero puede sobreajustarse a los datos.
- **Random Forest (rf)**: M谩s preciso y robusto, reduce el sobreajuste mediante la combinaci贸n de muchos 谩rboles.
- **Extra Trees (et)**: Similar a los random forests pero m谩s aleatorio, lo que puede llevar a una reducci贸n adicional en la varianza y un tiempo de entrenamiento m谩s r谩pido.
